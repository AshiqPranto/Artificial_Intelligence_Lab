```python
import numpy as np
import matplotlib.pyplot as plt

# Define a linear function: y = mx + c
def linear_function(x, m, c):
    return m * x + c

# Define a non-linear function: y = ax^2 + bx + c
def non_linear_function(x, a, b, c):
    return a * x**2 + b * x + c

# Generate some sample data
x_values = np.linspace(-5, 5, 100)  # Generate 100 equally spaced points between -5 and 5

# Parameters for linear function
m = 1
c = 3

# Parameters for non-linear function
a = 10
b = 2
c = 1

# Calculate y values for linear function
y_linear = linear_function(x_values, m, c)

# Calculate y values for non-linear function
y_non_linear = non_linear_function(x_values, a, b, c)

# Plotting the functions
plt.figure(figsize=(10, 5))

# Linear function plot
plt.subplot(1, 2, 1)
plt.plot(x_values, y_linear, label=f'Linear Function: y = {m}x + {c}', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
plt.legend()

# Non-linear function plot
plt.subplot(1, 2, 2)
plt.plot(x_values, y_non_linear, label=f'Non-linear Function: y = {a}x^2 + {b}x + {c}', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Non-linear Function')
plt.legend()

plt.tight_layout()
plt.show()

# Analyzing the functions
# For linear function, the relationship between x and y is a straight line.
# For non-linear function, the relationship between x and y is curved, indicating a non-linear relationship.
```
